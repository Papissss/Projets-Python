# Visualisation des Variables Quantitatives et univariées
### Il faut toujours procéder par une analyse univariée. Pour cela, la connaissance du type de donnée est fondamentale selon quelle soit quantitative,
### qualitative, continue ou discrète.
### Lors de l'analyse de variablees quantitatives univariées, il est essentiel de choisir le type de graphique pour une interprétation correcte des 
### des données.

## \** Histogramme\**
### - \** Usage : \** Il est idéal pour montrer la \***distribution\*** d'une variable quantitative.

### - \** Description : \** Il permet de visualiser la \*** fréquence\*** de chaque intervalle facilitant la compréhension de la distribution des 
### données.

## \** Diagramme en boîte (Boxplot) : \**

### C'est la boîte à moustache. 
### - \** Usage : \** Le diagramme en boîte est utilisé pour visualiser la \*** répartition \*** des données en terrmes de quartiles.
### - \** Description : \** Il illustre la \***médiane\***, les \***quartiles\*** et les \*** valeurs aberrantes \***.
### Il est utile pour identifier les \***anomalies\*** et \*** comparer les distributions \***.

## \** La courbe de densité \**

### - \** Usage : \** Parfait pour visualiser la distribution d'une \***variable continue\***.
### - \** Description : \** La courbe de densité est une version lissée de l'histogramme offrant une visualisation claire de la distribution.

# \** En Résumé \**

### - Variables quantitatives univariées : 

## - \*** Hitogramme\***
## - \*** Diagramme en boîte (Boxplot) encore appelé la boîte à moustache\***
## - \*** Courbe de densité (version lissée de l'histogramme) \***

```python
Package : matplotlib

Un package est une collection de modules regroupés dans un dossier. Il sert à organiser des fonctionnalités liées dans un espace cohérent.
Dans ce cas :

matplotlib est un package complet pour créer des graphiques en Python.
Il contient plusieurs modules qui permettent de tracer divers types de visualisations (2D et 3D).

Module : matplotlib.pyplot

Un module est un fichier Python qui contient des fonctions, des classes ou des variables utilisables dans un programme.

matplotlib.pyplot est un module spécifique du package matplotlib.
Il offre une interface simple et fonctionnelle, similaire à MATLAB, pour créer des graphiques en Python.

Rôle de matplotlib.pyplot

Le module matplotlib.pyplot contient des fonctions pour créer et personnaliser des graphiques. Voici ce qu'il permet de faire :

Créer différents types de graphiques :

- Courbes (graphiques linéaires).
- Barres (diagrammes en barres).
- Histogrammes.
- Nuages de points (scatter plots).
- Boxplots, pie charts, etc.
          
- Personnaliser les graphiques :
Ajouter des titres, des légendes, des étiquettes d'axes.
Modifier les couleurs, styles de lignes, tailles des points.
Ajuster les tailles des graphiques, polices de caractères, etc.
Afficher les graphiques :

Utilisation de la fonction plt.show() pour afficher un graphique dans une fenêtre ou un notebook.

import matplotlib.pyplot as plt

# Données
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Tracé du graphique
plt.plot(x, y, label="Croissance")
plt.title("Exemple de graphique")
plt.xlabel("Temps")
plt.ylabel("Valeurs")
plt.legend()

# Afficher le graphique
plt.show()

Résumé

Package (matplotlib) : Ensemble de modules pour créer des graphiques.
Module (pyplot) : Interface fonctionnelle et conviviale pour tracer et personnaliser facilement des visualisations.
plt : Nom fréquemment utilisé pour appeler matplotlib.pyplot après l'importation.
```
